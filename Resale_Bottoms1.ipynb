## üßπ 4. Data Cleaning

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
#clean up names
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('/', '_')
```

```python
#Standardize condition category

df['condition'] = df['condition'].str.lower().str.replace('is_', '')
```

```python
df.head()
```

```python
#clean up hashtags
df['hashtags'] = df['hashtags'].str.lower().str.replace("['", "").str.replace("'", "").str.replace(",", " ")
```

```python
#clean up traits
df['traits'] = df['traits'].str.lower().str.replace("{'", "").str.replace("':", " ").str.replace(
    "'", "").str.replace(",", " ").str.replace('}', '')
```

```python
#clean up shipping
df['shipping_options'] = df['shipping_options'].str.lower().str.replace("{'", "").str.replace("':", " ").str.replace(
    "'", "").str.replace(",", " ").str.replace('}', '')
```

```python
#clean price drops
df['price_drops'] = df['price_drops'].str.lower().str.replace("[", "").str.replace(",", " ").str.replace(
    ",", " ").str.replace("]", "")
```

## üìä Visualizations and Insights from the Resale Bottoms Dataset

In this section, we explore key trends and patterns using data visualizations. <br> These charts provide insights into brand popularity, pricing behavior, <br> seller performance,  and more within the resale market for branded bottoms.

## üìä Top Resale Brands - Listing Counts

```python
#top 10 brands
top_brands = df['brand'].value_counts().head(10).reset_index()
top_brands.columns = ['brand', 'count']
```

```python
plt.figure(figsize=(12,6))
#bar plot
sns.barplot(data=top_brands, x='brand', y='count', palette='flare')
#label settings
plt.xticks(rotation=45)
plt.title('Top 10 most listed Resale Brands')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#top 10 designer names
top_designer = df['designer_names'].value_counts().head(10).reset_index()
top_designer.columns = ['designer_names', 'count']
```

```python
plt.figure(figsize=(12,6))
#bar plot
sns.barplot(data=top_designer, x='designer_names', y='count', palette='flare')
#label settings
plt.xticks(rotation=45)
plt.title('Top 10 most listed Designer Names')
plt.xlabel('Designer Names')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

### üí° Insight

These charts give a quick glance at which brands are listed most frequently. <br> 
All pretty close by Just "brand" name with the exception of Acne Studios falling off <br>
However when we look at the chart with 'Designer Names' or vintage associated with brand <br>
Rick Owens and Saint Laurent Paris dominate

## üíµ Distribution of Resale Prices

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')
```

```python
df_filtered = df[df['price'] < 500]
```

```python
#histogram
plt.figure(figsize=(10,6))

sns.histplot(data=df_filtered, x='price', bins=50, kde=True, color='blue')
plt.xticks(ticks=range(0, 500, 50))
plt.title('Distribution of Resale Prices Under $500')
plt.xlabel('Price ($)')
plt.ylabel('Number of Listings')
plt.tight_layout()
plt.show()
```

```python
df_filtered2 = df[df['price'] >= 500 & (df['price'] <= 2000)]
```

```python
# Filter data between 500 and 3000
df_filtered2 = df[(df['price'] >= 500) & (df['price'] <= 3000)]

# Plot
plt.figure(figsize=(12, 6))
sns.histplot(data=df_filtered2, x='price', bins=30, kde=True, color='blue')
plt.xticks(ticks=range(500, 3001, 100))  # tick marks every $50
plt.title('Distribution of Resale Prices Above $500')
plt.xlabel('Price ($)')
plt.ylabel('Number of Listings')
plt.tight_layout()
plt.show()
```

üí° Insight

First chart is listings under 500, and it shows we primarly have listings around <br>
50 dollar rand and the 150 range. The second chart is listings above 500 and shows <br>
$500 is also one of our strong listing price ranges

```python
plt.figure(figsize=(10, 6))

sns.boxplot(data=df, x='condition', y='price', palette='pastel')

plt.title('Resale Price Distribution by Condition')
plt.xlabel('Condition')
plt.ylabel('Price ($)')
plt.xticks(rotation=15)
plt.ylim(0, 3000)  # Limit y-axis from 0 to 2500
plt.yticks(ticks=range(0, 3000, 200))
plt.tight_layout()
plt.show()
```

### üí° Insight

As we can see in the above graph, 'used' and 'worn <br>
are close in price listing. Possibly showing conditions <br>
are not accurately taken into account, or condition does not <br>
play a huge role between the two. As expected 'new' has the <br>
highest priced listings, with 'gently used' in the middle

```python
size_counts = df['size'].value_counts()
```

```python
top_sizes = size_counts.nlargest(12)

#plot
plt.figure(figsize=(8, 8))
plt.pie(top_sizes, labels=top_sizes.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('colorblind'))
plt.title('Distribution of Sizes in Resale Listings')
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x='seller_rating',
    y='follower_count',
    hue='location',
    palette='Set2',
    alpha=0.6
)

plt.title('Seller Rating vs. Follower Count by Location')
plt.xlabel('Seller Rating')
plt.ylabel('Follower Count')
plt.legend(title='Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
